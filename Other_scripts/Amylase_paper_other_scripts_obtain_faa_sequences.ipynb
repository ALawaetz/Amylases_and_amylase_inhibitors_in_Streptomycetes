```python
import pandas as pd
import os
import re
from Bio import SeqIO
from pathlib import Path
```

```python
### Obtain fasta sequences for all GH13 proteins
base_dir = Path('/Users/andreaslawaetz/Streptomyces/dbcan_results')

for folder in base_dir.iterdir():
    if folder.is_dir():
        accession = folder.name
        overview_path = folder / "overview_GH13.tsv" 
        if overview_path.exists():
            df = pd.read_csv(overview_path, sep = '\t', header = 0)
            gene_ids = set(df['Gene ID'].tolist())
            input_fasta = f'/Users/andreaslawaetz/Streptomyces/Angeliga_project/SCOG/faa_files/{accession}/ncbi_dataset/data/{accession}/protein.faa'
            output_fasta = f'{overview_path}.faa'
            with open(output_fasta, "w") as out_handle:
                for record in SeqIO.parse(input_fasta, "fasta"):
                    # The ID in record.id should match your gene IDs
                    if record.id in gene_ids:
                        SeqIO.write(record, out_handle, "fasta")
```
